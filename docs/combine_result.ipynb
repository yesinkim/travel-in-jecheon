```python
from difflib import SequenceMatcher

def normalize(s):
    if isinstance(s, str):
        return s.strip().lower()
    return ""


def is_same(a, b):
    return normalize(a) == normalize(b)


df_base["q_eq_gt"] = df_base.apply(lambda r: is_same(r["question"], r["ground_truth"]), axis=1)

df_rag["q_eq_gt"] = df_rag.apply(lambda r: is_same(r["question"], r["ground_truth"]), axis=1)


# ---------------------------
# 3) question vs ground_truth 유사도 스코어 추가
# ---------------------------
def sim(a, b):
    return SequenceMatcher(None, str(a), str(b)).ratio()


df_base["q_gt_sim"] = df_base.apply(lambda r: sim(r["question"], r["ground_truth"]), axis=1)
df_rag["q_gt_sim"] = df_rag.apply(lambda r: sim(r["question"], r["ground_truth"]), axis=1)

# ---------------------------
# 4) 모델 간 성능 비교(ROUGE·BERT·Judge 등)
# ---------------------------
cols_to_compare = ["rougeL", "bert_f1", "judge_accuracy", "judge_helpfulness"]

comparison = pd.DataFrame(
    {
        "metric": cols_to_compare,
        "base_model": [df_base[c].mean() for c in cols_to_compare],
        "rag_model": [df_rag[c].mean() for c in cols_to_compare],
    }
)

comparison["difference"] = comparison["rag_model"] - comparison["base_model"]

# ---------------------------
# 5) 핵심 출력용
# ---------------------------
print("=== Question-GT 비교 예시(상위 5개) ===")
print(df_base[["question", "ground_truth", "q_eq_gt", "q_gt_sim"]].head())

print("\n=== 모델별 스코어 비교 ===")
print(comparison)
```
